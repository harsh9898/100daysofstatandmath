## Python


```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

### Hexagonal binning and plotting numeric vs numeric data


```python
library(hexbin)
```

```python
data = pd.read_csv('/content/mtcars.csv')

ax = data.plot.hexbin(x='mpg', y='cyl',
                         gridsize=30, sharex=False, figsize=(5, 4))
ax.set_xlabel('MPG')
ax.set_ylabel('CYL')
```

```python
data = pd.read_csv('/content/airquality.csv')
data.head(10)
```

```python
ax = data.plot.hexbin(x='Ozone', y='Temp',
                         gridsize=30, sharex=False, figsize=(5, 4))
ax.set_xlabel('Ozone')
ax.set_ylabel('Temperature')
```

### Contour plot for Ozone vs Temperatue in air quality data set


```python
ax = sns.kdeplot(data.Ozone, data.Temp)
ax.set_xlabel('Ozone')
ax.set_ylabel('Temperature')
```

### Categorical and numeric data

```python
ax = data.boxplot(by='Month', column='Temp')
ax.set_xlabel('Month')
ax.set_ylabel('Temperature')
plt.suptitle(' ')
```

### Violin plot for month vs Ozone

```python
ax = sns.violinplot(data.Month, data.Ozone,
                    inner='quartile', color='blue')
ax.set_xlabel('Month')
ax.set_ylabel('Monthly Ozone value')
```
