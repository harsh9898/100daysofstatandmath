### Hexagonal binning and plotting numeric vs numeric data


```r
install.packages('hexbin')
```

```r
library(hexbin)
```

```r
data <- datasets::mtcars
ggplot(data, (aes(x=mpg, y=cyl))) +
  stat_binhex(color='white') +
  theme_bw() +
  scale_fill_gradient(low='white', high='black') +
  labs(x='MPG', y='CYL')
```

```r
data <- datasets::airquality
data %>% head(1)
```

```r

ggplot(data, (aes(x=Ozone, y=Temp))) +
  stat_binhex(color='white') +
  theme_bw() +
  scale_fill_gradient(low='white', high='black') +
  labs(x='Ozone', y='Temperature')
```

### Contour plot for Ozone vs Temperatue in air quality data set


```r
ggplot(data, aes(Ozone, Temp)) +
  theme_bw() +
  geom_point(alpha=0.1) +
  geom_density2d(color='blue') +
  labs(x='Ozone', y='Temperature')
```

### Categorical and numeric data

### Violin plot for month vs Ozone

```r
ggplot(data=datasets::airquality, aes(Day, Ozone)) +
  geom_violin() +
  labs(x='Day', y='Daily Ozone Value')
```
