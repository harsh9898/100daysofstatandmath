```r
library(dplyr)
library(tidyr)
library(ggplot2)
library(magrittr) # for %>% operator
#install.packages('corrplot')
library(corrplot)
```

### Correlation heat map for mtcars

```r
data <- datasets::mtcars
res <- cor(data)
round(res, 2)
```

### This is correlation heatmap based on correlation matrix. The size of ellipse tells the positive and negative correlations

```r
corrplot(cor(data), method='ellipse')
```

### Below is just the visually different version in case anyone just wants to see one part of the correlation since other half is just the repeatation

```r
corrplot(cor(data), type = "lower", order = "hclust",
         tl.col = "black", tl.srt = 45)
```
