```r
library(dplyr)
library(tidyr)
library(ggplot2)
library(magrittr) # for %>% operator
```

### Box plot for mtcars' wt column, in which we can see two data points are outside of the quantiles

```r
data <- mtcars

boxplot(data[['wt']], ylab='MPG')
```

### Frequency Table for mtcars' cylinder to get the frequency of each car

```r
breaks <- seq(from=min(data[['cyl']]), 
              to=max(data[['cyl']]))
cyl_freq <- cut(data[['cyl']], breaks=breaks, 
                right=TRUE, include.lowest=TRUE)
data['cyl_freq'] <- cyl_freq
table(cyl_freq)
```

### Histogram based on frequency table of cyl column

```r
hist(data[['cyl']], breaks=breaks)
```

### Density plot (smooth version of histogram)

```r
hist(data$cyl, freq=FALSE)
lines(density(data$cyl), lwd=3, col='blue')
```
